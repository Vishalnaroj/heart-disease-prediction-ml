**Use mean when:**

The data is *numeric*

The *distribution is symmetric* (not skewed)

There are *no significant outliers*


**Avoid mean if:**

The feature has outliers → mean gets dragged toward them and becomes misleading.



**Use median when:**

The data is *numeric*

The* distribution is skewed*

*There are outliers *→ median is robust and won't be affected much

**Use mode when:**

The *feature is categorical*

You want to fill with the most frequent category

example:

sex, chest_pain_type, or thal might be categorical.

Use **mode** **if there's a missing value** in those.
